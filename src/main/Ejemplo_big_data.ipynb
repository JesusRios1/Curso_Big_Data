```python
"""Invocamos las dependencias instaladas"""
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType 
from pyspark.sql.types import ArrayType, DoubleType, BooleanType
from pyspark.sql.functions import col,array_contains
```

```python
"""Inicializamos una sesion con Spark"""
spark = SparkSession.builder.appName('MyFirstSession').getOrCreate()
```

```python
"""Realizamos la lectura de un archivo .CSV"""
df = spark.read.option("header",True).csv("/content/sample_data/california_housing_test.csv")
"""Se imprime la estructura de los datos"""
df.printSchema()
```

```python
"""Se muestra el contenido del archivo en un sample(muestra 20 registros)"""
df.show(truncate=False)
```

```python
df1 = df.withColumn("longitude",col("longitude").cast("Integer"))
df1.show()
df1.printSchema()
```

```python
df1 = df.withColumnRenamed("longitude","long").show()
```

```python
df.sort('longitude')[['longitude']].show(10)
```

```python
from pyspark.sql import functions as F
```

```python
df.sort(F.desc('longitude'))[['longitude']].show(10)
```

```python
distinctDF = df.distinct()
print("Distinct count: "+str(distinctDF.count()))
distinctDF.show(truncate=False)
```
